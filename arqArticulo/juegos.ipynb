```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Conversi√≥n de respuestas cualitativas a valores num√©ricos
# Define un diccionario para mapear las respuestas a valores num√©ricos
mapeo_respuestas = {
    "Totalmente en desacuerdo": 1,
    "En desacuerdo": 2,
    "Algo en desacuerdo": 3,
    "Ni de acuerdo ni en desacuerdo": 4,
    "Algo de acuerdo": 5,
    "De acuerdo": 6,
    "Totalmente de acuerdo": 7,

}

```

```python

# Aplicar el mapeo
df_numerico = df.copy()
df_numerico.iloc[:, 1:] = df.iloc[:, 1:].applymap(lambda x: mapeo_respuestas.get(x, None))

# Asegurarse de que los datos sean num√©ricos
df_numerico.iloc[:, 1:] = df_numerico.iloc[:, 1:].apply(pd.to_numeric, errors="coerce")

# Manejar valores faltantes (opcional: usar promedio)
df_numerico.fillna(df_numerico.mean(), inplace=True)

# Calcular el promedio por afirmaci√≥n
df_numerico["Promedio"] = df_numerico.iloc[:, 1:].mean(axis=1)

# Preparar el mapa de calor
plt.figure(figsize=(12, 8))
ax = sns.heatmap(
    df_numerico.iloc[:, 1:-1].T,  # Transponer para que las afirmaciones est√©n en el eje vertical
    cmap="coolwarm",
    annot=True,
    fmt=".2f",
    linewidths=0.5,
    cbar_kws={"label": "Nivel promedio de acuerdo"}
)

# Configurar etiquetas y t√≠tulo
plt.title("Mapa de Calor: Nivel de Acuerdo Promedio por Afirmaci√≥n", fontsize=14)
plt.xlabel("Afirmaciones", fontsize=12)
plt.ylabel("Niveles de Acuerdo", fontsize=12)
plt.xticks(rotation=45, ha="right")
plt.tight_layout()

# Mostrar el gr√°fico
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Diccionario de mapeo de respuestas a valores num√©ricos
response_map = {
    "Totalmente en desacuerdo": 1,
    "En desacuerdo": 2,
    "Ni de acuerdo ni en desacuerdo": 3,
    "Algo de acuerdo": 4,
    "De acuerdo": 5,
    "Totalmente de acuerdo": 6
}

# Definir las categor√≠as y las afirmaciones que pertenecen a cada una
categories = {
    "Gr√°ficos": [
        "Creo que los gr√°ficos del juego se adaptan al estado de √°nimo o al estilo del juego."
    ],
    "Audio": [
        "Siento que el audio del juego (por ejemplo, efectos de sonido, m√∫sica) mejora mi experiencia de juego."
    ],
    "Interacci√≥n Social": [
        "Disfruto de la interacci√≥n social dentro del juego."
    ],
    "Controles": [
        "Considero que los men√∫s del juego son f√°ciles de usar."
    ],
    "Inmersi√≥n": [
        "Me siento separado del mundo exterior mientras juego."
    ]
}

# Funci√≥n para convertir las respuestas en valores num√©ricos
def convert_responses(df):
    return df.applymap(lambda x: response_map.get(x, None))

# Convertir las respuestas en valores num√©ricos
numeric_df = convert_responses(df)

# Calcular el promedio por categor√≠a
category_averages = {}
for category, questions in categories.items():
    category_columns = [col for col in numeric_df.columns if any(q in col for q in questions)]
    category_averages[category] = numeric_df[category_columns].mean(axis=1).mean()

# Crear el gr√°fico de l√≠neas
plt.figure(figsize=(10, 6))
plt.plot(list(category_averages.keys()), list(category_averages.values()), marker='o', color='b')

# Configurar el gr√°fico
plt.title("Tendencias de acuerdo seg√∫n categor√≠as de juego")
plt.xlabel("Categor√≠as")
plt.ylabel("Promedio de acuerdo")
plt.grid(True)
plt.tight_layout()

# Mostrar el gr√°fico
plt.show()
```
